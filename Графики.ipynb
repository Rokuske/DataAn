```python
#Импорт библиотек для графиков
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

```python
x=np.linspace(0, 10, 5) #Задать ось иксов
#np.linspace Возвращает равномерно распределенные числа за указанный интервал.
y = x**2 #Задать игрики 
```

```python
plt.figure() #plt.figure(figsize=(x, y)) — создать график размера  (x,y) ;
plt.plot(x,y)
"""
plt.plot(x, y, params) — 
нарисовать график по точкам с координатами из  x  по горизонтальной оси и из  y  по вертикальной оси.
Точки будут соединятся в том порядке, в котором они указаны в этих массивах;
"""
plt.show() #plt.show() — показать график;
```

```python
x = np.arange(0, 2*np.pi, 0.1) #Работает аналогично linspace только 3й аргумет задает шаг а не кол во точек
y = np.cos(x)
```

```python
plt.figure()
plt.title("График косинуса") #Установить название графика
plt.grid() #добавить сетку на график
plt.plot(x,y)
plt.show()
```

```python
#Нарисовать график по заданным точкам x и y
plt.figure()
plt.plot([0, 0.25, 1], [0, 1, 0.5])
plt.show()
```

```python
#Начертить x(t) и y(t)
t = np.linspace(0, 2*np.pi, 100)
plt.figure(figsize=(5,5))
plt.plot(np.tan(t), np.sin(t))
plt.show()
```

```python
"""
Для простой регулировки цветов и типов линий после пары  x  и  y
координат вставляется форматная строка. Первая буква определяет цвет
('r' — красный, 'b' — синий и т.д.),
дальше задаётся тип линии ('-' — сплошная, '--' — пунктирная, '-.' — штрих-пунктирная и т.д.)
"""
x =np.arange(0, 15, 0.5)
plt.figure(figsize=(15,10))
plt.plot(x, x**2, "b--")
plt.plot(x, x**3, "r-")
plt.show()
```

```python
#Если в качестве "типа линии" указано 'o', то это означает рисовать точки кружочками и не соединять их линиями; аналогично, 's' означает квадратики.
x = np.linspace(0, 1, 11)

plt.figure()
plt.plot(x, x ** 2, 'ro') #красные круги
plt.plot(x, 1 - x, 'gs') #зеленые квадраты
plt.show()
```

```python
x = np.linspace(0, 2*np.pi, 30)
plt.figure(figsize=(10,10))
plt.plot(x, np.sin(x), linewidth=2, color="r", dashes=[10, 4, 3, 1], label = "sin(x)")
 #НЕ МОГУ РАЗОБРАТЬСЯ
"""
plt.xticks(np.linspace(0, 2 * np.pi, 9),  # Где сделать отметки
           ('0',r'$\frac{1}{4}\pi$',r'$\frac{1}{2}\pi$',  # Как подписать
            r'$\frac{3}{4}\pi$',r'$\pi$',r'$\frac{5}{4}\pi$',
            r'$\frac{3}{2}\pi$',r'$\frac{7}{4}\pi$',r'$2\pi$'),
           fontsize=20)         
"""
plt.legend(fontsize=20, loc=0) #Разместить легенду в позиции loc и размеров fontsize
plt.show()
#Dashes=[]-Устанавливает вид линии и цыфри показывают длинну тире и длинну пропусков
#linewidth=x-толщина линии 
```

```python
#Упражение: постройте пунктирный график функции  y=x3 . Подпишите оси, заголовок и легенду.
x = np.arange(0, 10, 1)
y = x**3
```

```python
plt.figure()

plt.plot(x, y, dashes=[5,3], color="b", label="y=x^3")
plt.grid()
plt.legend(fontsize=10, loc=0)
plt.title("Пробный график")
plt.show()
```
