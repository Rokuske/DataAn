```python
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D # для отрисовки 3D проекции

matplotlib.rc("font", size=18) # для увеличения шрифта подписей графиков
```

```python
phones = pd.read_csv("phones.csv")
```

```python
phones.head(7)
```

```python
#Модель регрессии
```

```python
fig=plt.figure(figsize=(15,15))
ax = plt.axes()
ax.scatter(phones["disk"], phones["price"], s=100)
plt.xlabel("диск")
plt.ylabel("Цена")
plt.show()
```

```python
# импортируем модуль, отвечающий за линейную регрессию
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
#
x=pd.DataFrame(phones.disk)
y=pd.DataFrame(phones.price)
```

```python
model.fit(x,y)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
fig=plt.figure(figsize=(15,15))
ax = plt.axes()
plt.plot(x, model.predict(x), color="r", linewidth = 5)
ax.scatter(phones["disk"], phones["price"], s=100)
plt.xlabel("диск")
plt.ylabel("Цена")
plt.show()
```

```python
model.score(x,y)
```
